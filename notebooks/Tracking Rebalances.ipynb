# Tracking Rebalances in Bike Share Systems

Bike shares provide a flexible transit system to their customers.  Since you're just renting a bike for a small amount of time, you have the freedom to ride it to any one of the other stations in a city.  As long as there's a station near where you want to go, you can take whatever means you want to get there.  In former modes of public transit you have to plan your route carefully, noting directions and route transfers in order to get as close as possible to where you want to go.

In addition to this flexible mode of transportation, Bike Shares add additional bikers to the road, which has, according to some statistical observation, improved the level of bike safety as a whole in a given city.  As there are more bikers in a city, the number of bike and car related accidents and death goes down.

Fewer violent deaths, more exercise, and less of a dependency on burning gasoline to get you where you're going.

All that said, there's some unpredictable costs with running a Bike Share.  For the most part you can easily track how much wear and tear your bikes get, and with credit cards you can easily hold individual customers accountable for returning the bike that they rent.  Something that is much more difficult, however, is how many bikes to have in the system.  In a perfect world you have the same number of bikers to bikes.  An even harder problem is that every station needs as many bikes as it as bikers are going to walk up to it.

Throughout the day Nice Ride has trucks that move through the city relocating entire rows of bikes to correct these imbalances in their system.  Naturally with the ebb and flow of the day, more people commute to work in the morning and go home early evening.  We relocate across substantial distances just to do our jobs and then head back home for the night to get ready for the next day.

If you reduce the number of rebalances you need to do in a week, you reduce the cost of the system.  This is a system I want to work well, and reducing the costs to it can only improve it.

```python
# Okay, so now to get a single station's data
bikes.index = bikes["Start Station"]
sample_station = stations_by_size[stations_by_size > 1800].order().tail(1)
bikes_from_station = bikes.ix[sample_station.index]
bikes_from_station.index = bikes_from_station.start_ts
bikes_from_station.start_ts.resample("d", how="count").plot(figsize=(12,8), legend=True)
bikes.index = bikes["End Station"]
bikes_to_station = bikes.ix[sample_station.index]
bikes_to_station.index = bikes_to_station.start_ts
bikes_to_station.start_ts.resample("d", how="count").plot(figsize=(12,8))
```
